## VI Algorithm details - 

#### 1. Initialize V(s,a) with all zeros

#### 2. Calculate the Q function at state s and action a using - 
$$Q_i(s, a) = \sum_{s'} P(s' | s,a) \cdot [ r(s,a,s') + \gamma V_{i}(s')]$$

#### 3. Get the new value function using -
$$V_{(i+1)}(s) = \max_a \sum_{s'} P(s' | s,a) \cdot [ r(s,a,s') + \gamma V_{i}(s')] = \max_a Q_i(s,a)$$

#### Repeat until convergence
That's it for Value Iteration Algorithm

#### Now, get the optimal action at state s using the calculated values - 
 $$\pi^*(s) = argmax_a \sum_{s'} P(s' | s,a) \cdot [ r(s,a,s') + \gamma V_{i}(s')] = argmax_a Q_i(s,a)$$

```python
#Step 1
state_values = {s : 0 for s in mdp.get_all_states()}
```

```python
%%writefile mdp_get_action_value.py
#Visualize - Step 2
def get_action_value(mdp, state_values, state, action, gamma):
    """ Compute Q(s,a) """
    #Summation over all the next states
    q_func = 0
    for state_p in mdp.get_next_states(state, action):
        model_prob = mdp.get_transition_prob(state, action, state_p)
        reward = mdp.get_reward(state,action,state_p)
        q_func += model_prob*(reward+gamma*state_values[state_p])    
    return q_func    
```

```python
#Real - Step 2
def get_action_value(mdp, state_values, state, action, gamma):
    """ Compute Q(s,a) """
    #Summation over all the next states
    q_func = 0
    for state_p in mdp.get_next_states(state, action):
        model_prob = mdp.get_transition_prob(state, action, state_p)
        reward = mdp.get_reward(state,action,state_p)
        q_func += model_prob*(reward+gamma*state_values[state_p])    
    return q_func    
```

```python
#Step 3
def get_new_state_value(mdp, state_values, state, gamma):
    """ Compute next V(s) """
    if mdp.is_terminal(state):
        return 0
    #List over all the actions first
    q_actions = []
    for action in mdp.get_possible_actions(state):
        q_actions.append(get_action_value(mdp, state_values, state, action, gamma))
    #Get the max and return
    return np.max(q_actions)
```

```python
gamma = 0.9            # discount for MDP
num_iter = 100         # maximum iterations, excluding initialization
# stop VI if new values are this close to old values (or closer)
min_difference = 0.001

#Initialize V(s) again - just in case
state_values = {s: 0 for s in mdp.get_all_states()}

for i in range(num_iter):

    #Compute new state values
    new_state_values = {}
    for state in state_values:
        new_state_values[state] = get_new_state_value(mdp, state_values, state, gamma)
        
    # Compute difference
    diff = max(abs(new_state_values[s] - state_values[s])
               for s in mdp.get_all_states())
    print("iter %4i   |   diff: %6.5f   |   " % (i, diff), end="")
    print('   '.join("V(%s) = %.3f" % (s, v) for s, v in state_values.items()))
    state_values = new_state_values

    if diff < min_difference:
        print("Terminated")
        break
```

```python
#Visualize
if has_graphviz:
    display(plot_graph_with_state_values(mdp, state_values))
```

## Now that V(s) is calculated, get optimal action

```python
def get_optimal_action(mdp, state_values, state, gamma=0.9):
    """ Find optimal action """
    if mdp.is_terminal(state):
        return None

    #List over all the actions first
    q_actions = []
    for action in mdp.get_possible_actions(state):
        q_actions.append(get_action_value(mdp, state_values, state, action, gamma))
    #Get the max for the action and return the action string, not index
   
    return mdp.get_possible_actions(state)[np.argmax(q_actions)]
```

```python
#Visualize
if has_graphviz:
    try:
        display(plot_graph_optimal_strategy_and_state_values(mdp, state_values))
    except ImportError:
        raise ImportError("Run the cell that starts with \"%%writefile mdp_get_action_value.py\"")
```

```python
# Measure agent's average reward

s = mdp.reset()
rewards = []
for _ in range(10000):
    s, r, done, _ = mdp.step(get_optimal_action(mdp, state_values, s, gamma))
    rewards.append(r)
    
print("average reward: ", np.mean(rewards))

assert(0.85 < np.mean(rewards) < 1.0)
```
